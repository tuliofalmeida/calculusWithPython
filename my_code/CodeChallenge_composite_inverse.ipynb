```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

### Part 1

Exercise 1

part a
$$f(x) = 2x^2 -4 \\
g(x) = 7|x| + 3$$

```python
def fx(x):
    return 2*(x**2) - 4

def gx(x):
    return 7*(abs(x)) + 3

domain = np.linspace(-5, 5, 200)

plt.plot(domain, fx(domain), label='f(x) = 2x^2 - 4')
plt.plot(domain, gx(domain), label='g(x) = 7|x| + 3')
plt.plot(domain, fx(gx(domain)), label='f(g(x))')
plt.plot(domain, gx(fx(domain)),'.', label='g(f(x))')
plt.xlabel('x')
plt.ylabel('y = f(x), g(x)')
plt.xlim(-5, 5)
plt.ylim(-10, 50)
plt.legend(loc='lower right')
```

part b
$$f(x) = \sin (x) \\
g(x) = ln(x) \\
h(x) = 2x^2 +5 $$

```python
domain = np.linspace(-300, 300, 1000)

def fx(x):
    return np.sin(x)

def gx(x):
    return np.log(x)

def hx(x):
    return 2*(x**2) + 5

plt.plot(domain, fx(gx(hx(domain))), label='f(g(h(x)))')
# plt.plot(domain, gx(hx(fx(domain))), label='g(h(f(x)))')
# plt.plot(domain, hx(fx(gx(domain))), label='h(f(g(x)))')
# plt.plot(domain, fx(hx(gx(domain))), label='f(h(g(x)))')
# plt.plot(domain, gx(fx(hx(domain))), label='g(f(h(x)))')
# plt.plot(domain, hx(gx(fx(domain))), label='h(g(f(x)))')

plt.xlabel('x')
plt.ylabel('y = f(x), g(x), h(x)')
plt.legend(loc='upper right')
```

Exercise 2

$$
f(x) = ln(2x) \\
g(x) = e^x/2
$$

```python
def fx(x):
    return np.log(2*x)

def gx(x):
    return np.exp(x)/2

domain = np.linspace(.01, 5, 54)

plt.plot(domain, fx(domain), label='f(x) = ln(2x)')
plt.plot(domain, gx(domain), label='g(x) = (e^x)/2')
plt.plot(domain, fx(gx(domain)), label='f(g(x))')
plt.plot(domain, gx(fx(domain)), '.', label='g(f(x))')
plt.xlabel('x')
plt.ylabel('y = f(x), g(x)')
plt.ylim(-4,10)
plt.legend(loc='upper right')
```

### Part 3

Exercise 3 - using sympy

$$f(x) = \sin (x) \\
g(x) = ln(x) \\
h(x) = 2x^2 +5 $$

```python
x = sp.symbols('x')
fx = sp.sin(x)
gx = sp.log(x)
hx = 2*(x**2) + 5

fghx = fx.subs(x, gx.subs(x, hx))
display(Math('f(g(h(x))) = %s' % sp.latex(fghx)))

sp.plot(fghx, (x, -100, 100), title='y = f(g(h(x)))')
```
