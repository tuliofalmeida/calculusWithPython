```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

Exercise 1: Important trig funcitons in sympy

$$  \lim_{\phi \rightarrow 0} = \big[  \frac{\cos(\phi)-1}{\phi} \big] = 0$$

```python
phi = sp.symbols('phi')


fx = sp.sin(phi)/phi 
display(Math('f(x) = '+ sp.latex(fx)))
sp.plot(fx, (phi, -10, 10), ylabel="$y = f(\phi)$", xlabel="$\phi$", title="Piecewise function", show=True)

# find the limits

# find the limit of the function as phi approaches 0 from the left
limitAt0Left = sp.limit(fx,phi,0,'-')
display(Math(f'\\lim_{{phi \\to 0^-}} f(phi) = {limitAt0Left}'))

# find the limit of the function as phi approaches 0 from the right
limitAt0Right = sp.limit(fx,phi,0,'+')
display(Math(f'\\lim_{{phi \\to 0^+}} f(phi) = {limitAt0Right}'))
```
