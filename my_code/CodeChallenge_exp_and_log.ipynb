$$

Exercise 5

Demonstrate that the natrual log and exp are inverse

```python
x = sp.symbols('x')

sp.plot((sp.log(x),(x,-10,10)),
        (sp.exp(x),(x,-10,2)),
        (sp.log(sp.exp(x)),(x,-10,10)),
        (sp.exp(sp.log(x)),(x,.1,10)),
        axis_center=(-4,-4),
        xlim=[-4,4], ylim=[-4,4])
```

### Part 2

The rules

$$x^a \times x^b = x^{a+b} ~~~~~~ ln(a \times b) = ln(a) + ln(b)$$

$$\frac{x^a}{x^b} = x^{a-b} ~~~~~~ ln(a^b) = bln(a)$$ 

$$(x^a)^b = x^{ab} ~~~~~~ ln(\frac{a}{b}) = ln(a) - ln(b)$$

Exercise 1 - first rule adding powers

```python
x,a,b = sp.symbols('x a b')

eq = x**a * x**b
eq
```

```python
eq = sp.simplify(eq)
eq
```

```python
eq = x**a * x**b
eq.subs({a:1,b:2})
```

Exercise 4 - power of log

```python
a,b = 2,4

display(Math('ln(a \\times b) == ln(a) + ln(b)'))
print(np.log(a*b),np.log(a)+np.log(b),np.log(a*b)==np.log(a)+np.log(b))

a,b = np.random.rand(100),np.random.rand(100)

eq = np.log(a*b)-(np.log(a)+np.log(b))

plt.plot(eq,'-o')
plt.ylim(-.0000001,.0000001)
```

```python
a,b = 2,4

display(Math('ln(a^b) == bln(a)'))
print(np.log(a**b),b*np.log(a),np.log(a**b)==b*np.log(a))

a,b = np.random.rand(100),np.random.rand(100)

eq = np.log(a**b) - (b*np.log(a))

plt.plot(eq,'-o')
plt.ylim(-.0000001,.0000001)
```

```python
a,b = 2,4

display(Math('ln(\\frac{a}{b}) == ln(a) - ln(b)'))
print(np.log(a/b),np.log(a) - np.log(b),np.log(a/b)==np.log(a) - np.log(b))

a,b = np.random.rand(100),np.random.rand(100)

eq = np.log(a/b) - (np.log(a) - np.log(b))

plt.plot(eq,'-o')
plt.ylim(-.0000001,.0000001)
```
