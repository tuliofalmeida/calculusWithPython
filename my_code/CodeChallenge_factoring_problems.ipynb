```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

Exercise 1

$$  \lim_{x \rightarrow a} = \frac{c(x^2 - (a+b) x +ab)}{x^2 + (b-a)x-ab} $$
$$ a,b \in \{-4 \rightarrow 4\} $$ 
$$ c \in \{-1,1\} $$

$$  \lim_{x \rightarrow a} = \frac{c(x-a)(x-b)}{(x-a)(x+b)} $$

```python
x = sp.symbols('x')
rand = np.arange(-4,5,1)
a,b = np.random.choice(rand),np.random.choice(rand)
c = np.random.choice([-1,1])

fx = (c*(x**2 - (a+b)*x + (a*b)))/(x**2 + (b-a)*x -(a*b))
fx2 = (c*(x-a)*(x-b))/((x-a)*(x+b))
```

```python
sp.plot(fx,(x,-10,10),ylim=(-10,10),title='$f(x) = %s$' % sp.latex(fx))

# fx
display(Math(r'\lim_{{x \to a}} {}'.format(sp.latex(sp.limit(fx,x,a)))))

# fx2
display(Math(r'\lim_{{x \to a}} {}'.format(sp.latex(sp.limit(fx2,x,a)))))

display(Math(r'f(a) = %s' % sp.latex(fx.subs(x,a))))
display(Math(r'f(a) = %s' % sp.latex(fx2.subs(x,a))))
```

Exercise 2 - square roots and conjugates

$$  \lim_{x \rightarrow -a} = \frac{\sqrt{x+b} + c}{x + a} $$
$$ a \in \{-4 \rightarrow 4\} $$ 
$$ b \in a +\{0 \rightarrow 4\} $$ 
$$ c = \sqrt{b-a} $$

```python
x = sp.symbols('x')
rand = np.arange(-4,5,1)
rand2 = np.arange(0,5,1)
a = np.random.choice(rand)
b = a + np.random.choice(rand2)
c = -np.sqrt(b-a)

fx = (sp.sqrt(x+b) + c)/(x+a)
display(Math(r'f(x) = %s' % sp.latex(fx)))
```

```python
sp.plot(fx,(x,-10,10),ylim=(0,11),xlim = (-4,8),title='$f(x) = %s$' % sp.latex(fx))

# fx
display(Math(r'\lim_{{x \to a}} {}'.format(sp.latex(sp.limit(fx,x,-a)))))

print('a = ',-a)
display(Math(r'f(a) = %s' % sp.latex(fx.subs(x,-a))))
```

Exercise 3 - absolute values

$$  \lim_{x \rightarrow a} = \mid x-a\mid + b $$
$$  \lim_{x \rightarrow a} = \frac{x-a}{\mid x-a\mid} $$
$$ a,b \in \{-4 \rightarrow 4\} $$ 

```python
x = sp.symbols('x')
rand = np.arange(-4,5,1)
a,b = np.random.choice(rand),np.random.choice(rand)

fx = sp.Abs(x-a) + b
display(Math(r'f(x) = %s' % sp.latex(fx)))

fx2 = (x-a)/sp.Abs(x-a)
display(Math(r'f2(x) = %s' % sp.latex(fx2)))
```

```python
sp.plot(fx,(x,-10,10),ylim=(-5,11),xlim = (-4,8),title='$f(x) = %s$' % sp.latex(fx))

display(Math(r'\lim_{{x \to a}} {}'.format(sp.latex(sp.limit(fx,x,a)))))

print('a = ',a)
display(Math(r'f(a) = %s' % sp.latex(fx.subs(x,a))))
```
