### Part 1

Exercise 1

Create a $2^{nd}$ order polynomial with random integer coefficients [-5,5]

$$ 
f(x) = a_1 + a_2x + a_3x^2
$$

```python
np.random.seed(0)

a1 = np.random.randint(-5, 5)
a2 = np.random.randint(-5, 5)
a3 = np.random.randint(-5, 5)
```

```python
x = np.linspace(-4,4,1000)
fx = a1 + a2*x + a3*x**2
plt.plot(x,fx)
x = 0
fx = a1 + a2*x + a3*x**2
plt.plot(x,fx,'.')
x = -1
fx = a1 + a2*x + a3*x**2
plt.plot(x,fx,'.')
x = 1
fx = a1 + a2*x + a3*x**2
plt.plot(x,fx,'.')
plt.title(f'$f(x) = {a1} + {a2}x + {a3}x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
```

Exercise 2 

Create a cosine-scaled exponential with random interger coefficient [-3,3] excluding 0

$$ f(\theta) = ae^{-\theta}\cos(2\pi\theta)$$

```python
theta = sp.symbols('theta')
a = np.random.randint(-3, 3,100)[np.random.randint(1,100)]
# np.random.choice([i for i in range(-3,4) if i!=0])

fx = a*sp.exp(-theta) * sp.cos(2*sp.pi*theta)
display(fx)
```

```python
sp.plot(fx, (theta, 0, 10), ylabel='f(theta)', xlabel='theta')
```

```python
# convert fx to plot using numpy
f = sp.lambdify(theta, fx, 'numpy')
x = np.linspace(-3,3,100)
plt.plot(x,f(x))
plt.plot(0,f(0),'.')
```

### Part 2

Exercise 3 

Random chose between the conditions

$$ f(x) =  \begin{cases}  \{x,x^2,x^3\}  & , & x \leq 0  \\
           \{ln(x), \sqrt{x} \} & , & x > 0 \end{cases} $$

```python
def first_cond(x,cond):
    return [np.power(x,1),np.power(x,2),np.power(x,3)][cond]

def second_cond(x,cond):
    return [np.log(x),np.sqrt(x)][cond]

x = np.linspace(-3,3,100)

cond1 = np.random.choice(3)
cond2 = np.random.choice(2)

y1 = [first_cond(i,cond1) for i in x if i <= 0]
y2 = [second_cond(i,cond2) for i in x if i > 0]

label1 = ['x','x**2','x**3']
label2 = ['log(x)','sqrt(x)']

plt.plot(x[x<=0],y1,label=label1[cond1])
plt.plot(x[x>0],y2,label=label2[cond2])
plt.title('Random piece-wise function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
