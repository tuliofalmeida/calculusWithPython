```python
import numpy as np
import matplotlib.pyplot as plt
```

Exercise 1

$y = x^2 + 3x^3 - x^4$

```python
x_domain = np.arange(-2,2.1,.1)
y_range = np.power(x_domain,2) + (np.power(x_domain,3)*3) - np.power(x_domain,4)
```

```python
plt.plot(x_domain,y_range)
plt.legend(['$y = x^2 + 3x^3 - x^4$'])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

Exeercise 2, using sympy

$f(\beta) = -\beta^4 + 3\beta^3 + \beta^2$


```python
import sympy as sp
```

```python
beta = sp.symbols('beta')
f_beta = np.power(beta,2) + np.power(beta,3)*3 - np.power(beta,4)

sp.plot(f_beta,(beta,-2,2),title='$f(\\beta) = -\\beta^4 + 3\\beta^3 + \\beta^2$')
```

Exercise 3

Convert sympy to numpy

```python
f_beta_np = sp.lambdify(beta,f_beta,'numpy')
domain_np = np.arange(-2,2.1,.1)
range_np = f_beta_np(domain_np)

plt.plot(domain_np,range_np)
plt.legend(['$f(\\beta) = -\\beta^4 + 3\\beta^3 + \\beta^2$'])
plt.xlabel('$\\beta$')
plt.ylabel('$f(\\beta)$')
plt.grid()
plt.show()
```
