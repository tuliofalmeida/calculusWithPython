```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

### Part 1

Exercise 1 - replicate the graph

$$f(x) =\begin{cases}0 & when & x < 0\\-2x & when & 0 \leq x < 3 \\ .1x^3 & when & x \geq 3 \end{cases}  $$

my solution

```python
general_domain = np.linspace(-3,5,1000)

sub_domain = np.zeros(len(general_domain))

for point in range(len(general_domain)):
    if general_domain[point] < 0:
        continue
    elif 0 <= general_domain[point] < 3:
        sub_domain[point] = -2*general_domain[point]
    else:
        sub_domain[point] = .1*general_domain[point]**3

plt.plot(general_domain, sub_domain)
```

mike solution

```python
general_domain = np.linspace(-3,5,1000)

piece1 = 0  * (general_domain < 0)
piece2 = -2 * general_domain * ((0 <= general_domain) & (general_domain < 3))
piece3 = .1 * general_domain**3 * (3 <= general_domain)

y = piece1 + piece2 + piece3

plt.plot(general_domain, y)
```

Exercise 2 - draw separate lines

$$f(x) =\begin{cases}0 & when & x < 0\\-2x & when & 0 \leq x < 3 \\ .1x^3 & when & x \geq 3 \end{cases}  $$

my solution

```python
general_domain = np.linspace(-3,5,1000)

sub_domain1 = np.zeros(len(general_domain)) * np.nan
sub_domain2 = np.zeros(len(general_domain)) * np.nan
sub_domain3 = np.zeros(len(general_domain)) * np.nan

for point in range(len(general_domain)):
    if general_domain[point] < 0:
        sub_domain1[point] = 0
    elif 0 <= general_domain[point] < 3:
        sub_domain2[point] = -2*general_domain[point]
    else:
        sub_domain3[point] = .1*general_domain[point]**3

plt.plot(general_domain, sub_domain1)
plt.plot(general_domain, sub_domain2)
plt.plot(general_domain, sub_domain3)
plt.legend(["Piece 1", "Piece 2", "Piece 3"])
plt.ylabel("$y = f(x)$")
plt.xlabel("$x$")
plt.title("Piecewise function")
plt.show()
```

mike solution

```python
gd = np.linspace(-3,5,1000)

plt.plot(gd[gd < 0], np.zeros(np.sum(gd < 0)), label="Piece 1")
plt.plot(gd[(0 <= gd) & (gd < 3)], -2*gd[(0 <= gd) & (gd < 3)], label="Piece 2")
plt.plot(gd[3 <= gd], .1*gd[3 <= gd]**3, label="Piece 3")
plt.legend()
plt.ylabel("$y = f(x)$")
plt.xlabel("$x$")
plt.title("Piecewise function")
plt.show()
```

Exercise 3 - piecewise in sympy

$$f(x) =\begin{cases}0 & when & x < 0\\-2x & when & 0 \leq x < 3 \\ .1x^3 & when & x \geq 3 \end{cases}  $$

```python
x = sp.symbols('x')
fx = sp.Piecewise((0, x < 0), (-2*x, (0 <= x) & (x < 3)), (.1*x**3, 3 <= x))
display(Math('f(x) = '+ sp.latex(fx)))
sp.plot(fx, (x, -3, 5), ylabel="$y = f(x)$", xlabel="$x$", title="Piecewise function", show=True)
```

### Part 2

Exercise 4 - Evaluate the funcitona t or near a point

$$f(x) =\begin{cases}0 & when & x < 0\\-2x & when & 0 \leq x < 3 \\ .1x^3 & when & x \geq 3 \end{cases}  $$

```python
# find the value of the function at x = .5 
tarrget_y = .5

# using sympy

x = sp.symbols('x')
fx = sp.Piecewise((0, x < 0), (-2*x, (0 <= x) & (x < 3)), (.1*x**3, 3 <= x))
display(Math('sympy: f(.5) = ' + sp.latex(fx.subs(x, tarrget_y))))

# using numpy 

general_domain = np.linspace(-3,5,100) # I changed the amount of points to reach the same value

sub_domain = np.zeros(len(general_domain))

for point in range(len(general_domain)):
    if general_domain[point] < 0:
        continue
    elif 0 <= general_domain[point] < 3:
        sub_domain[point] = -2*general_domain[point]
    else:
        sub_domain[point] = .1*general_domain[point]**3

# shift the plot then the value closest to 0 is the value at x = .5 (target_y)
# then the minima will be the value of the function at x = .5 (target_y)
idx = np.argmin(np.abs(general_domain - tarrget_y))

display(Math('numpy: f(.5) =' + str (np.round(sub_domain[idx],3))))
```
