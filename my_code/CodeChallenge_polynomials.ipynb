```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

Exercise 1

```python
x = sp.symbols('x')
a = np.random.rand(4)
y = a[0]+ a[1]*np.power(x,1) + a[2]*np.power(x,2) - a[3]*np.power(x,3)
```

```python
domain_np = sp.lambdify(x,y,'numpy')
range_y = domain_np(np.arange(-5,6,.1))
```

```python
plt.plot(np.arange(-5,6,.1),range_y)
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.show()
```

other solution

```python
coefs = np.random.randn(4)

x = np.linspace(-5,5,101)
y = np.zeros(len(x))

fname = '$f(x) = '
for i,c in enumerate(coefs):
    y += c*x**i
    fname += '+ '[int(c<0)] + f'{c:.2f}x^{i}'

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title(fname + '$')
plt.show()
```

Solution

```python
order = 10
x = np.linspace(-2*np.pi,2*np.pi,100)
y = np.zeros(len(x))

for n in range(1,order+1):
    sign = (-1)**(n-1)
    num = x**(2*n-1)
    den = np.math.factorial(2*n-1)
    this_term = sign*num/den

    plt.plot(x,this_term,'--')

    y+= this_term

plt.plot(x[::5],y[::5],'-o',color = 'k')
plt.xlim(x[[0,-1]])
plt.ylim(-5,5)
```
