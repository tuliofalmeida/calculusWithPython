```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

### Part 1

Exercise 1

$$ f = \frac{x^3}{3} + 100 \sqrt{\mid  x \mid } $$

$$  \lim_{x \rightarrow 5}(cf) = c  \big(\lim_{x \rightarrow 5}(f)\big) $$




```python
x = sp.symbols('x')
c = np.random.rand()
fx = (x**3)/3 + 100 * sp.sqrt(sp.Abs(x))
fxx = c*((x**3)/3 + 100 * sp.sqrt(sp.Abs(x)))

# plot
f = sp.lambdify(x, fx, 'numpy')
ffx = sp.lambdify(x, fxx, 'numpy')
xVals = np.linspace(-5,5,100)
yVals = f(xVals) * c

plt.plot(xVals, yVals, label='c * lim(f(x))')
plt.plot(xVals, ffx(xVals), '--', label='lim(c*f(x))')
plt.title('$c * f(x) = x^3/3 + 100*sqrt(|x|)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

Exercise 2 

$$ f = ln(x) + x^2 $$
$$ g = \exp(-x) + x^3 $$
$$ \lim_{x \rightarrow \pi}(f +g) =  \big(\lim_{x \rightarrow \pi}f\big) + \big(\lim_{x \rightarrow \pi}g\big)$$

```python
x = sp.symbols('x')
fx = sp.log(x) + x**2
gx = sp.exp(-x) + x**3

print('lim(f(x) + g(x)):')
display(sp.N(sp.limit(fx + gx, x, sp.pi)))
print('lim(f(x)) + lim(g(x)):')
display(sp.N(sp.limit(fx, x, sp.pi)) + sp.N(sp.limit(gx, x, sp.pi)))
```

### Part 2

Exercise 3

$$ f = ln(x) + x^2 $$
$$ g = \exp(-x) + x^3 $$
$$ \lim_{x \rightarrow \pi}(f \times g) =  \big(\lim_{x \rightarrow \pi}f\big) \times \big(\lim_{x \rightarrow \pi}g\big)$$
$$ \lim_{x \rightarrow \pi}(f^3) =  \big(\lim_{x \rightarrow \pi}f\big) \times \big(\lim_{x \rightarrow \pi}f\big) \times \big(\lim_{x \rightarrow \pi}f\big)$$

```python
x = sp.symbols('x')
fx = sp.log(x) + x**2
gx = sp.exp(-x) + x**3

print('lim(f(x) * g(x)):')
display(sp.N(sp.limit(fx * gx, x, sp.pi)))
print('lim(f(x)) * lim(g(x)):')
display(sp.N(sp.limit(fx, x, sp.pi)) * sp.N(sp.limit(gx, x, sp.pi)))
print()
print('lim(f(x))**3:')
display(sp.N(sp.limit(fx, x, sp.pi))**3)
print('lim(f(x))*lim(f(x))*lim(f(x)):')
display(sp.N(sp.limit(fx, x, sp.pi))*sp.N(sp.limit(fx, x, sp.pi))*sp.N(sp.limit(fx, x, sp.pi)))
```

Exercise 4

$$ f = ln(x) + x^2 $$
$$ g = \exp(-x) + x^3 $$
$$ h = x^3 + x^2 + x $$
$$ \lim_{x \rightarrow \pi}(f / g) =  \big(\lim_{x \rightarrow \pi}f\big) / \big(\lim_{x \rightarrow \pi}g\big)$$
$$ \lim_{x \rightarrow 0}(g / h) =  \big(\lim_{x \rightarrow 0}g\big) / \big(\lim_{x \rightarrow 0}h\big)$$

```python
x = sp.symbols('x')
fx = sp.log(x) + x**2
gx = sp.exp(-x) + x**3
hx = x**3 + x**2 + x

print('lim(f(x)/g(x)):')
display(sp.N(sp.limit(fx /gx, x, sp.pi)))

print('lim(f(x)) / lim(g(x)):')
display(sp.N(sp.limit(fx, x, sp.pi))/sp.N(sp.limit(gx, x, sp.pi)))

print()

print('lim(g(x)/h(x)):')
display(sp.N(sp.limit(gx /hx, x, 0)))
print('lim(g(x)) + lim(h(x)):')
display(sp.N(sp.limit(gx, x, 0)) / sp.N(sp.limit(hx, x, 0)))
```
