```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

Exercise 1

- Implement and plot the func $f(\theta) = \frac{\sin \theta}{\theta}$ in sympy
- Implement and plot $g(\theta) = \frac{-1}{\theta}$ in sympy
- Implement and plot $h(\theta) = \frac{1}{\theta}$ in sympy
- Do the same plots in matplotlib
- Compute and print the limits in sympy

```python
theta = sp.symbols('theta') 

f_theta = sp.sin(theta)/theta
g_theta = -1/theta
h_theta = 1/theta

sp.plot(f_theta,g_theta,h_theta, (theta, -100, 100), ylabel='f(θ)', nb_of_points=1000,adaptive = False, ylim=(-1,1))

display(Math(r'\lim_{{\theta \to \infty}} f(\theta) = {}'.format(sp.limit(f_theta, theta, sp.oo))))
display(Math(r'\lim_{{\theta \to \infty}} g(\theta) = {}'.format(sp.limit(g_theta, theta, sp.oo))))
display(Math(r'\lim_{{\theta \to \infty}} h(\theta) = {}'.format(sp.limit(h_theta, theta, sp.oo))))
```

```python
# lambdify those functions
f = sp.lambdify(theta, f_theta, modules=['numpy'])
g = sp.lambdify(theta, g_theta, modules=['numpy'])
h = sp.lambdify(theta, h_theta, modules=['numpy'])

x = np.linspace(-100, 100, 1000)

plt.plot(x, f(x), label='$f(θ) = sin(θ)/θ$')
plt.plot(x, g(x), label='$g(θ) = -1/θ$')
plt.plot(x, h(x), label='$h(θ) = 1/θ$')

plt.title('Plot of $f(θ) = sin(θ)/θ$, $g(θ) = -1/θ$ and $h(θ) = 1/θ$')
plt.legend()
plt.ylim(-1,1)
plt.show()

plt.plot(x, np.sign(f(x)), label='$sign(f(θ) = sin(θ)/θ)$')
plt.plot(x, np.sign(g(x)), label='$sign(g(θ) = -1/θ)$')
plt.plot(x, np.sign(h(x)), label='$sign(h(θ) = 1/θ$)')

plt.title('Plot of $sign(f(θ) = sin(θ)/θ)$, $sign(g(θ) = -1/θ)$ and $sign(h(θ) = 1/θ$')
plt.legend()
plt.show()
```

Exercise 2

- Implement the function $f0 = f1 \cdot f2$ in sympy
$$ f0 = \frac{\tan(\theta)}{\theta} $$
$$ f1 = \frac{\sin(\theta)}{\theta} $$
$$ f2 = \frac{1}{\cos(\theta)} $$ 
- Plot using matplotlib
- Compute functions values at $\theta = 0$
- Compute and print the limits in sympy as $\theta \longrightarrow 0$

```python
theta = sp.symbols('theta')

f0 = sp.tan(theta)/theta
f1 = sp.sin(theta)/theta
f2 = 1/sp.cos(theta)
f3 = f1 * f2

print('Limits')
display(Math(r'\lim_{{\theta \to 0}} \frac{{tan(\theta)}}{{\theta}} = {}'.format(sp.limit(f0, theta, 0))))
display(Math(r'\lim_{{\theta \to 0}} \frac{{sin(\theta)}}{{\theta}} = {}'.format(sp.limit(f1, theta, 0))))
display(Math(r'\lim_{{\theta \to 0}} \frac{{1}}{{cos(\theta)}} = {}'.format(sp.limit(f2, theta, 0))))
display(Math(r'\lim_{{\theta \to 0}} \frac{{sin(\theta)}}{{\theta}} \cdot \frac{{1}}{{cos(\theta)}} = {}'.format(sp.limit(f3, theta, 0))))

f0 = sp.lambdify(theta, f0, modules=['numpy'])
f1 = sp.lambdify(theta, f1, modules=['numpy'])
f2 = sp.lambdify(theta, f2, modules=['numpy'])
f3 = sp.lambdify(theta, f3, modules=['numpy'])

x = np.linspace(-5,5,1000)

plt.plot(x, f0(x), label='$f(θ) = tan(θ)/θ$',linewidth=2)
plt.plot(x, f1(x), label='$f(θ) = sin(θ)/θ$',linewidth=1)
plt.plot(x, f2(x), label='$f(θ) = 1/cos(θ)$', linestyle='--')
plt.plot(x, f3(x), label='$f(θ) = sin(θ)/θ \cdot 1/cos(θ)$', linestyle='--',linewidth=.75)

plt.title('Plot of $f(θ) = tan(θ)/θ$, $f(θ) = sin(θ)/θ$, $f(θ) = 1/cos(θ)$ and $f(θ) = sin(θ)/θ \cdot 1/cos(θ)$')
plt.ylim(-20,20)
plt.xlim(-3,3)
plt.legend()
plt.show()

display('Functions values for θ = 0')
display(Math(r'f(0) = \frac{{tan(0)}}{{0}} = {}'.format(f0(0))))
display(Math(r'f(0) = \frac{{sin(0)}}{{0}} = {}'.format(f1(0))))
display(Math(r'f(0) = \frac{{1}}{{cos(0)}} = {}'.format(f2(0))))
display(Math(r'f(0) = \frac{{sin(0)}}{{0}} \cdot \frac{{1}}{{cos(0)}} = {}'.format(f3(0))))
```

Exercise 3

- Implement and plot the function in sympy $f(\theta) = \theta^2 \exp^{-\theta^2}\ln(\theta^2)\sin(\theta)$
- Compute and print the value aht $\theta = 0$ and the limit as $\theta \longrightarrow 0$
- Determine the domain of $f(\theta)$
- Compute the limits of each part (as $\theta \longrightarrow 0$), and the product of those limits

```python
theta = sp.symbols('theta')

f_theta = theta**2*sp.exp(-theta**2)*sp.ln(theta**2)*sp.sin(theta)

sp.plot(f_theta, (theta, -10, 10), ylabel='f(θ)', nb_of_points=1000,adaptive = False)

print("Limit at 0")
display(Math(r'\lim_{{\theta \to 0}} \theta^2 e^{{-\theta^2}} ln(\theta^2) sin(\theta) = {}'.format(sp.limit(f_theta, theta, 0))))

f = sp.lambdify(theta, f_theta, modules=['numpy'])
print('Value at 0')
display(Math(r'f(0) = \theta^2 e^{{-\theta^2}} ln(\theta^2) sin(\theta) = {}'.format(f(0))))

print('Determine the domain')
display(sp.calculus.util.continuous_domain(f_theta, theta, sp.S.Reals))

print("Separated limits")
display(Math(r'\lim_{{\theta \to 0}} \theta^2 = {}'.format(sp.limit(theta**2, theta, 0))))
display(Math(r'\lim_{{\theta \to 0}} e^{{-\theta^2}} = {}'.format(sp.limit(sp.exp(-theta**2), theta, 0))))
display(Math(r'\lim_{{\theta \to 0}} ln(\theta^2) = {}'.format(sp.limit(sp.ln(theta**2), theta, 0))))
display(Math(r'\lim_{{\theta \to 0}} sin(\theta) = {}'.format(sp.limit(sp.sin(theta), theta, 0))))
```
