```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from IPython.display import display, Math
```

Exercise 1

```python
domain = np.arange(-np.pi,2*np.pi,0.1)

plt.plot(domain, np.sin(domain))
plt.plot(domain, np.sin(domain)**2)
plt.plot(domain, np.sin(domain**2))
```

Exercise 2

```python
plt.plot(domain, np.sin(np.cos(domain)))
plt.plot(domain, np.cos(np.sin(domain)))
plt.plot(domain, np.cos(domain))
```

```python
domain2 = np.arange(-3,6,0.1)

plt.plot(domain2, np.sin(domain2))
plt.plot(domain2, np.cos(domain2))
```
