Exercise 2 - The idea is not todo limits but understand the concept

$$  \lim_{x \rightarrow 1} f(x) =~~? $$

$$ Start~~with~~x0 = [a-1,a+1] \\
for i iterations:
Evaluate~~and~~store~~y = f(x0) \\

Update~~x0 \leftarrow(x0+a)/2 $$


```python
a = 1

x0 = np.array([a-1,a+1])

iterations = 10
limits = np.zeros((iterations,2))
xvals = np.zeros((iterations,2))

for i in range(iterations):
    xvals[i,:] = x0
    limits[i,:] = np.cos(x0**2)**2 + np.pi

    x0 = (x0+a)/2

print('Limits from the right:')
for i in range(iterations):
    display(Math('\\lim_{{x \\rightarrow 1}} f({}) = {}'.format(xvals[i,0], limits[i,0])))

print('Limits from the right:')
for i in range(iterations):
    display(Math('\\lim_{{x \\rightarrow 1}} f({}) = {}'.format(xvals[i,1], limits[i,1])))
```

```python
# print out in a table
print('Limit from the left:')
print(np.vstack((xvals[:,0],limits[:,0])).T)

print(' ')
print('Limit from the right:')
print(np.vstack((xvals[:,1],limits[:,1])).T)

# print(' ')
# print(f'Function value at x={a}')
# print(fx(a))
```
